#3. Perform text preprocessing on the headlines using NLTK functions such as removing stop words, tokenization, stemming, and lemmatization. 

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
#Stopwords Removal
```

```python
# Initialize NLTK stopwords
stop_words = set(stopwords.words('english'))
```

```python
# Define a function for text preprocessing to remove stop words
def remove_stopwords(text):
    # Tokenize the text
    words = nltk.word_tokenize(text)

    # Remove stop words
    filtered_words = [word for word in words if word.lower() not in stop_words]

    # Join the filtered words back into a sentence
    preprocessed_text = ' '.join(filtered_words)

    return preprocessed_text

# Apply the preprocessing function to the 'headlines' column in the DataFrame
df['stopword_headline'] = df['headline'].apply(remove_stopwords)

# Display the DataFrame with preprocessed headlines
df.head()
```

```python
#Stemming
```

```python
import pandas as pd
import nltk
from nltk.stem.snowball import SnowballStemmer
```

```python
# Initialize the Snowball Stemmer for English
s_stemmer = SnowballStemmer(language='english')
```

```python
# Define a function for text preprocessing with stemming
def stem_text(text):
    # Tokenize the text
    words = nltk.word_tokenize(text)

    # Stem each word using the Snowball Stemmer
    stemmed_words = [s_stemmer.stem(word) for word in words]

    # Join the stemmed words back into a sentence
    preprocessed_text = ' '.join(stemmed_words)

    return preprocessed_text
# Apply the preprocessing function to the 'headlines' column in the DataFrame
df['stemmed_headline'] = df['stopword_headline'].apply(stem_text)

# Display the DataFrame with stemmed headlines
df.head()
```

```python
#Lemmatization
```

```python
# Initialize the WordNet Lemmatizer
lemmatizer = WordNetLemmatizer()

# Define a function for text preprocessing with lemmatization
def lemmatize_text(text):
    # Tokenize the text
    words = nltk.word_tokenize(text)

    # Lemmatize each word
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]

    # Join the lemmatized words back into a sentence
    preprocessed_text = ' '.join(lemmatized_words)

    return preprocessed_text

# Apply the preprocessing function to the 'headlines' column in the DataFrame
df['lemmatized_headline'] = df['stopword_headline'].apply(lemmatize_text)

# Display the DataFrame with lemmatized headlines
df.head()
```

#4. Perform Part-of-Speech (POS) tagging on the preprocessed headlines using NLTK's POS tagger.

```python
import pandas as pd
import nltk
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from nltk.tokenize import word_tokenize
```

```python
# Define a function for POS tagging
def pos_tagging(text):
    # Tokenize the text
    words = word_tokenize(text)

    # Perform POS tagging
    pos_tags = pos_tag(words)

    return pos_tags

# Apply POS tagging to the 'lemmatized_headline' column
df['pos_tags'] = df['lemmatized_headline'].apply(pos_tagging)

# Display the DataFrame with POS tags
df.head()
```

#3. Preprocess the headlines using spaCy functions such as removing stop words, tokenization, stemming, and lemmatization. 

```python

# Define a function for text preprocessing to remove stop words
def remove_stopwords(text):
    # Process the text using spaCy
    doc = nlp(text)
    
    # Remove stop words and join the remaining tokens
    filtered_tokens = [token.text for token in doc if not token.is_stop]
    preprocessed_text = ' '.join(filtered_tokens)

    return preprocessed_text

# Assuming you already have your DataFrame loaded
# Apply the preprocessing function to the 'headline' column in the DataFrame
df2['stopword_spacy'] = df2['headline'].apply(remove_stopwords)

# Display the DataFrame with preprocessed headlines
df2.head()
```

```python
# Define a function to perform lemmatization
def lemmatize_text(text):
    doc = nlp(text)
    lemmatized_text = " ".join([token.lemma_ for token in doc])
    return lemmatized_text

# Apply lemmatization to a specific column (e.g., 'headline_column_name')
df2['lemmatized_headlines'] = df2['stopword_spacy'].apply(lemmatize_text)

# print the DataFrame with lemmatized headlines
df2.head()
```

#4. Perform Part-of-Speech (POS) tagging on the preprocessed headlines using spaCy's POS tagger. 

```python
# Define a function for POS tagging
def pos_tagging(text):
    # Process the lemmatized text using spaCy
    doc = nlp(text)
    
    # Extract POS tags and join them with the corresponding tokens
    pos_tags = [f"{token.text} ({token.pos_})" for token in doc]
    pos_tagged_text = ' '.join(pos_tags)

    return pos_tagged_text

# Apply POS tagging to the lemmatized text
df2['pos_tagged_headline'] = df2['lemmatized_headlines'].apply(pos_tagging)

# Display the DataFrame with preprocessed, lemmatized, and POS-tagged headlines
df2.head()
```

#5. Perform Named Entity Recognition (NER) on the preprocessed headlines using spaCy's NER tagger. 

```python
def ner_tagging(text):
    # Process the preprocessed text using spaCy
    doc = nlp(text)
    
    # Extract named entities and their labels
    named_entities = [(ent.text, ent.label_) for ent in doc.ents]
    
    return named_entities

# Apply NER tagging to the preprocessed text
df2['ner_entities'] = df2['lemmatized_headlines'].apply(ner_tagging)

# Display the DataFrame with preprocessed and NER-tagged headlines
df2.head()
```

#6. Utilize spaCy's pipeline object to efficiently perform text processing tasks. 

```python
import pandas as pd
import spacy

# Load the spaCy English language model and create a spaCy pipeline
nlp = spacy.load("en_core_web_sm")

# Enable the 'tagger' component in the pipeline if it's not already enabled
if not nlp.has_pipe("tagger"):
    tagger = nlp.create_pipe("tagger")
    nlp.add_pipe(tagger)

# Define a function to preprocess a DataFrame using spaCy's pipeline
def preprocess_dataframe(df):
    # Process the entire DataFrame using spaCy's pipe method
    processed_data = list(nlp.pipe(df['headline'], disable=["parser"]))

    # Extract stop word-removed, lemmatized text and POS tags
    preprocessed_headlines = [" ".join([token.lemma_ for token in doc if not token.is_stop]) for doc in processed_data]
    pos_tagged_headlines = [" ".join([f"{token.text} ({token.pos_})" for token in doc]) for doc in processed_data]

    # Enable NER and extract named entities
    nlp.enable_pipe("ner")
    ner_entities = [list(doc.ents) for doc in processed_data]

    return preprocessed_headlines, pos_tagged_headlines, ner_entities

# Assuming you already have your DataFrame loaded
# Apply the preprocessing function to the DataFrame
preprocessed_headlines, pos_tagged_headlines, ner_entities = preprocess_dataframe(df2)

# Add the preprocessed data to the DataFrame
df2['preprocessed_headlines'] = preprocessed_headlines
df2['pos_tagged_headlines'] = pos_tagged_headlines
df2['ner_entities'] = ner_entities

# Display the DataFrame with all text processing tasks applied
df2.head()
```

#7. Use scikit-learn's TfidfVectorizer and LinearSVC in a pipeline for text vectorization and classification. 

```python
#Performing Sentimental Analysis for framing the target variable.
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
```

```python
def review_rating(string):
    # Initialize the VADER sentiment analyzer
    sid = SentimentIntensityAnalyzer()

    # Get sentiment scores
    sentiment_scores = sid.polarity_scores(string)

    # Determine sentiment based on the compound score
    compound_score = sentiment_scores["compound"]

    if compound_score >= 0.05:
        return "Positive"
    elif compound_score <= -0.05:
        return "Negative"
    else:
        return "Neutral"
```

```python
# Test the function on your review above:
df2['rating'] = df2['headline'].apply(review_rating)
```

```python
df2.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X = df2['lemmatized_headlines']
y = df2['rating']

#split data into train and test

X1_train, X1_test, y1_train, y1_test = train_test_split(X, y, test_size=0.2, random_state=11)
```

```python
text_clf = Pipeline([
    ('tfidf', TfidfVectorizer()),  
    ('clf', LinearSVC())
])

#Fit the pipeline to the training data

text_clf.fit(X1_train, y1_train)
# Predict the test cases
y1_pred = text_clf.predict(X1_test)

# Print the overall accuracy
print(f"Overall Accuracy: {accuracy_score(y1_test, y1_pred):.2f}")

```

```python
#create a confusion matrix
cm = confusion_matrix(y1_test, y1_pred)
print(cm)
```

```python
# Print the classification report
print(classification_report(y1_test, y1_pred))
```
